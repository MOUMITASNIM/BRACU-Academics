```python
#task_6
s1=input("Enter:")
s2=s1.split(", ")
n_list=[]
for i in s2:
  n_list.append(int(i))

large_n=n_list[0]
count=0
for n2 in n_list:
  if(n2>large_n):
    large_n=n2
    count+=1
print("My list:",s2) 
print("Largest number in the list is",large_n,"which was found at index",count,".")
```

```python
#task_8
list_one = [1, 2, 3, 4, 5, 6, 7, 8, 9]
list_two = [10, 11, 12, -13, -14, -15, -16]
new_list = []
for element in list_one:
    if element % 2 == 0:
        new_list.append(element)
for element1 in list_two:
    if element1 % 2 == 0:
        new_list.append(element1)
print(new_list)
```

```python
#task_9
data = input("Enter:")
data = data + " "
output = []
final = []
empty_str = ""
for value in range(0,len(data)):
    if data[value] != " ":
        empty_str += data[value]
    else:
        element = int(empty_str)
        output.append(element)
        empty_str = ""
        #print(empty_str)
print("Original list:",output)
for item in output:
    if (item % 2 != 0):
        final.append(item)
print("Modified list:", final)
```

```python
#TASK_10
s=input("Enter:")
list_1 = s.split(",")
list_2 = []
list_3 = []
for i in list_1:
    data=int(i)
    list_2.append(data)
print("Input list:",list_2)
for element in list_2:
    if element not in list_3:
        list_3.append(element)

print("Modified list:", list_3)
```
