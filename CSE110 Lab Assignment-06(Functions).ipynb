```python
#task_13
def function_opp(opp,num1,num2):
    if opp == '+':
        output = num1 + num2
    elif opp == '-':
        output = num1 - num2
    elif opp == '/':
        output = num1/num2
    elif opp == '*':
        output = num1*num2
    return output
print(function_opp(input('Operator: '),float(input("First number: ")),float(input("Second number: "))))
```

```python
#task_14
def removing_char(string, position):
    str_1 = string[0]
    rem = " "
    for i in range(1, len(string)):
        if (i % position == 0):
            rem = rem + string[i]
        else:
            str_1 = str_1 + string[i]
    str_1 = str_1 + rem

    return str_1


n = input()
number = int(input())
print(removing_char(n, number))
```

```python
#task_15
def grocery_app(order_items,location="Dhanmondi"):
    cost = {"Rice": 105, "Potato": 20, "Chicken": 250, "Beef": 510, "Oil": 85}
    total_c = 0
    for element in order_items :
        total_c += cost[element]
    if location == "Dhanmondi":
        total_c += 30
    else:
        total_c += 70
    return total_c


order_items = [element for element in input()[1:-1].replace('"', "").replace(" ", "").split(",")]
location = input()

# Function Call
if len(location) > 0:
    print(grocery_app(order_items,location))
else:
    print(grocery_app(order_items))
```
