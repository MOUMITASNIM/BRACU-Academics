```python
#task_2
def selection(my_list):
  for i in range(len(my_list)):
    min=my_list[i]
    minIdx=i
    for j in range(i+1,len(my_list)):
      if(min>my_list[j]):
        min=my_list[j]
        minIdx=j
    my_list[minIdx],my_list[i]=my_list[i],my_list[minIdx]
  return my_list

#driver code
print(selection([10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]))
```

```python
#task_5
def my_function(lists):
    for i in range(len(lists)-1):
        for j in range(0,len(lists)-i-1):
            if lists[j]<lists[j+1]:
                temp=lists[j+1]
                lists[j+1]=lists[j]
                lists[j]=temp
    return lists

lst = [ ["Alan", 95, 87, 91], ["Turing", 92, 90, 83], ["Elon", 87, 92, 80], ["Musk", 85, 94, 90] ]
cse110=[]
phy111=[]
mat110=[]
for i in lst:
    cse110.append(i[1])
    phy111.append(i[2])
    mat110.append(i[3])
a=my_function(cse110)
b=my_function(phy111)
c=my_function(mat110)
def stu_list(x,y):
    student_list=[]
    for i in x:
        for j in lst:
            if i==j[y]:
                student_list.append(j[0])
    return student_list
subject=input()
subject=subject.lower()
if subject=="cse110":
    for i in stu_list(a,1):
        print(i)
elif subject=="phy111":
    for k in stu_list(b,2):
        print(k)
elif subject=="mat110":
    for s in stu_list(c,3):
        print(s)
else:
    print("Error")
```

```python
#task_7
list_one = input('list_one= ')[1:-1].split(",")
list_two = input('list_two= ')[1:-1].split(",")
list_one1, list_two1, temp = [], [], 0
for iter1 in list_one:
  list_one1.append(int(iter1.strip()))
for iter2 in list_two:
  list_two1.append(int(iter2.strip()))

mergedList = list_one1 + list_two1

for iter3 in range(len(mergedList)):
  for iter4 in range(len(mergedList)-iter3-1):
    if mergedList[iter4] > mergedList[iter4+1]:
      temp = mergedList[iter4]
      mergedList[iter4] = mergedList[iter4+1]
      mergedList[iter4+1] = temp

print("Sorted list =", mergedList)

if len(mergedList) % 2 == 0:
  median = (mergedList[(len(mergedList)//2)-1] + mergedList[(len(mergedList)//2)]) / 2
elif len(mergedList) % 2 == 1:
  median = mergedList[len(mergedList)//2]

print("Median =", median)
```

```python
#task_8
list_one = input('Enter a list: ')[1:-1].split(",")
list_one1 = []
for iter1 in list_one:
  list_one1.append(int(iter1.strip()))
sum = 0
sumList, num1, num2 = [], [], []

for iter2 in range(len(list_one1)-1):
  for iter3 in range(iter2 + 1, len(list_one1)):
    sum = list_one1[iter2] + list_one1[iter3]
    if sum >= 0:
      sumList.append(sum)
    else:
      sumList.append(sum*(-1))
    num1.append(list_one1[iter2])
    num2.append(list_one1[iter3])

for iter4 in range(len(sumList)):
  for iter5 in range(len(sumList)-iter4-1):
    if sumList[iter5] > sumList[iter5+1]:
      temp1, temp2, temp3 = sumList[iter5], num1[iter5], num2[iter5]
      sumList[iter5] = sumList[iter5+1]
      sumList[iter5+1] = temp1
      num1[iter5] = num1[iter5+1]
      num1[iter5+1] = temp2
      num2[iter5] = num2[iter5+1]
      num2[iter5+1] = temp3

print("Two pairs which have the smallest sum =", num1[0], "and", num2[0])
```
