Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

```python
#You can run this driver code cell to understand the methods of Stack class
st = Stack()
st.push(4)
st.push(3)
st.push(5)
st.push(1)
st.push(9)

print('Peeked Element: ',st.peek())
print('Popped Element: ',st.pop())
print('Popped Element: ',st.pop())
print('Popped Element: ',st.pop())
print('Peeked Element: ',st.peek())
print('Popped Element: ',st.pop())
print('Popped Element: ',st.pop())
print('Peeked Element: ',st.peek())
print('Popped Element: ',st.pop())
print(st.isEmpty())
```

You can print your stack using this code segment

```python
def print_stack(st):
  if st.isEmpty():
    return
  p = st.pop()
  print('|',p,end=' ')
  if p<10:
    print(' |')
  else:
    print('|')
  #print('------')
  print_stack(st)
  st.push(p)

# st = Stack()
# st.push(4)
# st.push(3)
# st.push(5)
# st.push(1)
# st.push(9)
# print_stack(st)
# print('------')
```

***TASK : Parenthesis Balancing***

```python
def balance_parenthesis(string):

  st=Stack()

  for x in string:
    if x in "({[":
      st.push(x)

    elif x == ')':
      a=st.pop()
      if  a!='(':
        return False
    elif x == '}':
      a=st.pop()
      if  a!='{':
        return False
    elif x == ']':
      a=st.pop()
      if  a!='[':
        return False

  if st.isEmpty():
    return True
  else:
    return False

a = '{))1+2*(3/4)}'
value = balance_parenthesis(a)
if value:
  print('Balanced')
else:
  print('Unbalanced')
```
