```python
def reverse_Matrix(matrix):
    rv=np.zeros((len(matrix),len(matrix[0])),dtype=int)
    for i in range(len(matrix)) :
        for j in range(len(matrix[0])):
            rv[i][j]=matrix[-i-1][-j-1]
    return rv


import numpy as np
matrix = np.array([
[14,  8,  0,  4],
[9,  8,  13,  13],
[9,  3,  1,  4],
[2,  10,  13,  6]
])
print_matrix(matrix)
print('################')
print()
reversed_matrix = reverse_Matrix(matrix)
print_matrix(reversed_matrix)

#This should print
#|  6  |  13 |  10 |  2  |
#-------------------------
#|  4  |  1  |  3  |  9  |
#-------------------------
#|  13  |  13  |  8 |  9 |
#-------------------------
#|  4 |  0  |  8  |  14  |
#-------------------------
```

Task 5: Matrix Compression

```python
def compress_matrix(mat):

  row,colum=matrix.shape
  arr=np.zeros((2,2),dtype=int)

  for i in range(row):
    for j in range(colum):
      if i<row//2:
        if j<colum//2:
          arr[0][0]+=matrix[i][j]
        else:
          arr[0][1]+=matrix[i][j]
      else:
        if j<colum//2:
          arr[1][0]+=matrix[i][j]
        else:
          arr[1][1]+=matrix[i][j]

  return arr


matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

Task 6: Game Arena

```python
import numpy as np
def play_game(arena):

   pnt=0
   ch_P=np.array([[0,1],[0,-1],[1,0],[-1,0]])

   for i in range(len(arena)) :
       for j in range(len(arena[0])):
           if arena[i][j] >= 50 and arena[i][j]%50==0:
             for x in range(len(ch_P)) :
               if  i+[x][0]>=0 and j+ch_P[x][1]>=0 :
                 if i+ch_P[x][0]<=len(arena)-1 and j+ch_P[x][1]<=len(arena[0])-1 :

                    new=arena[i+ch_P[x][0]][j+ch_P[x][1]]
                    if new == 2 :
                      pnt+=2

   if pnt <10:
      print(f"Points Gained: {pnt}. Your team is out.")
   if pnt >=10:
      print(f"Points Gained: {pnt}. Your team has survived the game.")
   return None



import numpy as np

arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.

print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```

Bonus Task: Primary vs Secondary Diagonal

```python
def check_diagonal(matrix1, matrix2):


  m=np.array([0]*len(matrix1))
  n=np.array([0]*len(matrix1))

  for i in range(len(matrix1)):
      for j in range(len(matrix1[0])) :
          if i == j:
            m[i]= matrix1[i][j]
          if i+j == len(matrix1)-1:
            n[i]= matrix2[i][j]


  for i in range(len(m)):
      if m[i] != n[i] :
        return "NO"
  return "YES"

array1 = np.array([[0, 4, 1], [7, 2, 5], [3, 6, 0]])
array2 = np.array([[3, 6, 0], [5, 2, 7], [0, 4, 1]])

print(check_diagonal(array1, array2)) #This should print YES
print(".............")
array1 = np.array([[0, 9, 9, 1], [9, 0, 2, 9], [9, 3, 0, 9], [4, 9, 9, 0]])
array2 = np.array([[4, 9, 9, 0], [9, 0, 3, 9], [9, 0, 2, 9], [0, 9, 5, 1]])

print(check_diagonal (array1, array2)) #This should print NO
```
