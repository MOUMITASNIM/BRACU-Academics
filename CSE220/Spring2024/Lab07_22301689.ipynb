```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def inorder_predecessor(root, x):

    if not root:                    #conditions for checking empty root
        return None

    Predecessor1 = None            # Initializing as None

    while root:                    # Traversing the tree until reaching the node with value less than x

        if root.elem >= x.elem:    # If current node value is greater than or equal to x, moves to its left child
            root = root.left

        else:
            Predecessor1 = root     # If current node value is less than x, updates predecessor and moves to its right child
            root = root.right

    return Predecessor1             #Return the last node that was modified before the predecessor


#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
```

```python
def swap_child(root, level, M):

  if root==None:                                      #checking if the root is empty
    return None


  new_n = BTNode(root.elem)                           # Creating a new node with the same element as the current node.

  if level<M:
    new_n.left  = swap_child(root.right,level+1,M)    # Recursively swap children for the right subtree.
    new_n.right = swap_child(root.left,level+1,M)     # Recursively swap children for the left subtree.

  else:
    new_n.left  = swap_child(root.left,level+1,M)
    new_n.right = swap_child(root.right,level+1,M)


  return new_n



#Driver Code
root=BTNode('A')

#Write other nodes by yourself from the given tree of Doc File

root = BTNode('A')
root.left = BTNode('B')
root.right = BTNode('C')
root.left.left = BTNode('D')
root.left.right = BTNode('E')
root.right.right = BTNode('F')
root.right.right.left = BTNode('J')
root.left.left.left = BTNode('G')
root.left.left.right = BTNode('H')
root.left.right.left = BTNode('I')


print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root)   #Given Tree Inorder Traversal: G D H B I E A C J F
print()

root2 = swap_child(root, 0, 2)
print('Swapped Tree Inorder Traversal: ', end = ' ')
inorder(root2)  #Swapped Tree Inorder Traversal: J F C A I E B G D H
```
