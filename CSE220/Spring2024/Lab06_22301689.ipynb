```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    if self.search_hashtable(s) == 'Found':
      print(s,'Already Inserted. Cannot reinsert.')
      print('===============================')
      return
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'(key: {head.key}, value: {head.value})', end = '-->')
        head = head.next
      print('None')
      print()
      idx += 1


  #Do it by yourself
  def __hash_function(self, key):

    if len(key) % 2 != 0:              # Checking if the length of the key = odd  or not
      key += 'Num'                     # add "Num" to the key string as character
    summ = 0
    for i in range(0, len(key), 2):  # Iterating through the key two characters at once

          summ += int(str(ord(key[i])) + str(ord(key[i+1]))) #Concatenate the two characters' ASCII values and add them to the "summ".

    return summ % len(self.ht)       # Return the modulus of the summation with the size of the hashtable
                                     #calculation of hashtable



  #Do it by yourself

  def search_hashtable(self, s):

    hashed_index = self.__hash_function(s[0]) #Calculatng the hashed index of the key

    head = self.ht[hashed_index]   #Gets the head of the linked list ==>> index in the hashtable

    while head != None:            #iterating through the linked list

      if head.key == s[0]:        #checking if current node's key corresponds to the tuple's 1st element
        if head.value == s[1]:    #checking if the current node's value corresponds to the 2nd element of the tuple "s" only if the key matches.

          return 'Found'          # If the key and value of the current node match s, return 'Found'

      head = head.next            # Shifting to the next node

                                  # If it reaches to the last point of the list without finding "s",
    return 'Not Found'            #then returns 'Not Found'
```

Implement the __hash_function() and remove() methods

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):   #adds each key-value pair from an array to the hash table.
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'({head.key}, {head.value})', end = '-->')
        head = head.next
      print('None')
      idx += 1


  #Do it by yourself
  def __hash_function(self, key):  #determines where to place the key in the hash table.


    return (key + 3) % 6     #computes the hash value associated with a given key.inserting 3 to the key,divides the result by 6. This implies that,
                             #for a hash table of size six, the hash value remains constant between 0 and 5.

  #Do it by yourself
  def remove(self, key):    #removes a key-value pair from the hash table

        hashed_index = self.__hash_function(key)
        head = self.ht[hashed_index]


        if head is None:  #if head is empty, returns None
            return


        if head.key == key:           # Remove the head node (checking if the key matches the  head)
            self.ht[hashed_index] = head.next
            return


        prev = head                  # Traversing the list and removing the node if found
        while head != None:
            if head.key == key:
                prev.next = head.next
                return
            prev = head
            head = head.next
```
