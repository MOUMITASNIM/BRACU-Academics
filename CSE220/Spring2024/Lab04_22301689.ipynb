**Assignment Part 1: Doubly Linked List**

For Assignment Part 1, you should write your full code in the following cells along with your driver codes on your own.

```python
#Assignment Part 1

class Patient:
    def __init__(self, id, name, age, bloodgroup):
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup

class DoublyNode:
    def __init__(self, elem, next=None, prev=None):
        self.elem = elem
        self.next = next
        self.prev = prev
```

```python
class WRM:
    def __init__(self):
        self.head = DoublyNode(None)
        self.head.next = self.head
        self.head.prev = self.head

    def registerPatient(self, id, name, age, bloodgroup):
        new_patient = Patient(id, name, age, bloodgroup)
        new_node = DoublyNode(new_patient)

        new_node.next = self.head.next
        new_node.prev = self.head
        self.head.next.prev = new_node
        self.head.next = new_node

    def servePatient(self):
        if self.head.next == self.head:
            return None

        served_patient = self.head.next.elem
        self.head.next = self.head.next.next
        self.head.next.prev = self.head

        return served_patient

    def cancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head

    def canDoctorGoHome(self):
        return self.head.next == self.head

    def showAllPatient(self):
        current_node = self.head.next
        while current_node != self.head:
            print(current_node.elem.name)
            current_node = current_node.next

    def reverseTheLine(self):
        current = self.head.next
        while current != self.head:
            current.next, current.prev = current.prev, current.next
            current = current.prev
        self.head.next, self.head.prev = self.head.prev, self.head.next
wrm = WRM()
```

```python
#Write a Tester Code in this cell
print("**Welcome to Waiting Room Management System**")

while True:
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Reverse The Line")
    print("7. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        bloodgroup = input("Enter patient blood group: ")

        wrm.registerPatient(id, name, age, bloodgroup)
        print("Patient added successfully.")

    elif choice == 2:
        patient = wrm.servePatient()
        if patient:
            print("Patient being served:", patient.name)
        else:
            print("No patient to serve.")

    elif choice == 3:
        print("All patients in sequence to serve:")
        wrm.showAllPatient()

    elif choice == 4:
        if wrm.canDoctorGoHome():
            print("No one is waiting. Hence,yes, doctor can go home.")
        else:
            print("There are patients out there. Hence, no, doctor cannot go home.")

    elif choice == 5:
        wrm.cancelAll()
        print("All appointments cancelled till next input.")

    elif choice == 6:
        wrm.reverseTheLine()
        print("Line order reversed properly.")

    elif choice == 7:
        print("Exiting Waiting Room Management System")
        break
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  count=0
  stack=Stack()
  open=["<"]
  close=[">"]
  bracket= {">" : "<"}

  for m in string:
    if m in open:
      stack.push(m)
    elif m in close :
         x=stack.pop()
         if bracket[m]==x :
            count+=1
  return count


print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):

    a   = 0
    new_s = Stack()


    while not stack.isEmpty():
        new_s.push(stack.pop())
        a += 1

    output = Stack()

    for i in range(a):
        if i != a-n :
            output.push(new_s.pop())
        else:
            new_s.pop()

    st2= Stack()

    while not output.isEmpty():
          st2.push(output.pop())
    while not st2.isEmpty():
          stack.push(st2.pop())



print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack):
  st = Stack()

  while not stack.isEmpty():

    x = stack.pop()
    y = stack.peek()

    if x != y:
      st.push(x)
  return st

print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
