Task: DNA Synthesizing

You are a bioinformatics researcher working on synthesizing DNA sequences from multiple columns of nitrogen-bases. Each column represents a set of related samples.

Your task is to implement a function synthesize_sequence(given_samples) that takes a 2D array of nitrogen-bases as input and right rotates the given 2D array by 1 unit and returns it.

Sample Input 1:
given_samples = ( [ ['A', 'C', 'G', 'G'],
                                 ['T', 'G', 'A', 'A'],
                                 ['G', 'G', 'A', 'C'],
                                 ['C', 'T', 'G', 'A'] ] )
Output 1:
                                 [ 'G', 'A', 'C', 'G'],
                                 [ 'A', 'T', 'G', 'A'],
                                 [ 'C', 'G', 'G', 'A'],
                                 [ 'A', 'C', 'T', 'G']
Sample Input 2:
given_samples = ( [ ['A', 'C', 'G', 'G'],
                                 ['T', 'G', 'A', 'A'],
                                 ['A', 'G', 'A', 'C'],
                                 ['A', 'T', 'G', 'A'] ] )
Output 2:
                                 ['G', 'A', 'C', 'G'],
                                 ['A', 'T', 'G', 'A'],
                                 ['C', 'A', 'G', 'A'],
                                 ['A', 'A', 'T', 'G']

```python
import numpy as np

def printArray(givenArray):
    rows, columns = givenArray.shape
    for row in range(rows):
        for column in range(columns):
            print(givenArray[row][column], end=" ")
        print()

def synthesize_sequence(given_sequence):
    num_rows, num_cols = given_sequence.shape
    new_seq = np.empty_like(given_sequence)  # Create an empty array of the same shape

    for col in range(num_cols):
        new_seq[:, (col + 1) % num_cols] = given_sequence[:, col]  # Right rotate by 1

    return new_seq

# Driver code
given_sequence = np.array([['A', 'C', 'G', 'G'],
                           ['T', 'G', 'A', 'A'],
                           ['G', 'G', 'A', 'C'],
                           ['C', 'T', 'G', 'A']])

print("Original Array")
printArray(given_sequence)
print("=====================")
synthesizedArray1 = synthesize_sequence(given_sequence)
print("After Synthesizing")
printArray(synthesizedArray1)
print("=====================")

given_sequence = np.array([['A', 'C', 'G', 'G'],
                           ['T', 'G', 'A', 'A'],
                           ['A', 'G', 'A', 'C'],
                           ['A', 'T', 'G', 'A']])

print("Original Array")
printArray(given_sequence)
print("=====================")
synthesizedArray2 = synthesize_sequence(given_sequence)
print("After Synthesizing")
printArray(synthesizedArray2)
print("=====================")
```
