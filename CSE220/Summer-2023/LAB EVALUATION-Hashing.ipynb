```python
class Book:
  def __init__(self,name,next=None):
    self.name=name
    self.next=next

class Library:
  def __init__(self, size):
    self.shelves = [None]*size


  def add_book(self, s):
    if self.find_book(s) == 'Found':
      print(s,'Already added')
      print('===============================')
      return
    key = self.find_shelf(s)
    node = Book(s)
    if self.shelves[key] == None:
      self.shelves[key] = node
    else:
      node.next = self.shelves[key]
      self.shelves[key] = node

  def create_from_array(self, arr):
    for i in arr:
      self.add_book(i)


  def print_All_books(self):

    for i in self.shelves:
        while i:
          print(i.name,end="- > ")
          i=i.next
        print(None)


 #Do by yourself
  def find_shelf(self, s):
    shelf_number = None
    ch_count = 0
    vow_count = 0
    vowel = ["a","e","i","o","u","A","E","I","O","U"]
    for i in s:
      ch_count += 1
      if i in vowel:
        vow_count += 1

    shelf_number = (ch_count + vow_count) % len(self.shelves)
    return shelf_number

  def find_book(self, s):
    key=self.find_shelf(s)
    temp=self.shelves[key]
    while temp!=None:
      if temp.name==s:
        return 'Found'
      temp=temp.next
    return 'Not found'
```
