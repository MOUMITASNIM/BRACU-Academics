```python
class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    #Your code here
   if self.__top is None:
            return None
   else:
            popped = self.__top
            self.__top = self.__top.next
            popped.next = None
            return popped.elem


  def peek(self):
    if self.__top == None:
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```
