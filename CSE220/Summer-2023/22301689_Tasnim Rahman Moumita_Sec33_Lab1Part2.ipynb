```python
#part 2
#(Playing with Array and Probability Theory) 

import math

m=[10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]
def mean_calc(m):
    sum=0
    for i in range(len(m)):
        sum=sum+m[i]
    mean=sum/len(m)
    return mean

def sample_standard_div(m):
    sum=0
    l=len(m)-1
    for i in range(len(m)):
        sum+=math.pow((m[i]-mean_calc(m)),2)
    si=math.sqrt(sum * (1/l))
    return si

def standard_deviations(m):
    de_array=[None]*len(m)
    temp=0
    for i in range(len(m)):
        if(abs(m[i]-mean_calc(m)) >= (sample_standard_div(m)*1.5)):
            de_array[temp]=m[i]
            temp+=1
    n_array=[None]*temp
    if(temp>1):
      for i in range(0,temp):
        n_array[i]=de_array[i]
      return n_array
    else:
      print("[None]")

print(mean_calc(m))    
print(sample_standard_div(m))
print(standard_deviations(m))
```
