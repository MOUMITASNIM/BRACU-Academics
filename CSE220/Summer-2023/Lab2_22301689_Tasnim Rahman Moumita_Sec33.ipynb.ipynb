```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
  tile=floor.shape

  for i in range(tile[0]):
      if i % 2 == 0:
        for j in range(0,tile[1],2):
          print(floor[i][j],end=' ')
      else:
        for j in range(tile[1] -1,0,-2):
          print(floor[i][j],end=' ')
      print()

floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
import numpy as np
def landscape(m):
  land_s=m.shape
  n_array=np.zeros((land_s[1],land_s[0]),dtype=int)
  for i in range(land_s[0]):
    for j in range(land_s[1]):
       n_array[j][i]=m[i][j]

  return n_array





m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```
