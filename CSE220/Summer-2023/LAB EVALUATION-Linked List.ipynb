```python
class Node:
    def __init__(self, elem, next=None):
        self.elem, self.next = elem, next

def removeAllOdd(head):
    if head is None:
        return None

    # Check and remove the first node if it's odd
    while head is not None and head.elem % 2 == 1:
        head = head.next

    current = head

    # Traverse the linked list and remove odd nodes
    while current is not None and current.next is not None:
        if current.next.elem % 2 == 1:
            current.next = current.next.next
        else:
            current = current.next

    return head

def createList(arr):
    head = Node(arr[0])
    tail = head
    for i in range(1, len(arr)):
        newNode = Node(arr[i])
        tail.next = newNode
        tail = newNode
    return head

def printLinkedList(head):
    temp = head
    while temp != None:
        if temp.next != None:
            print(temp.elem, end='-->')
        else:
            print(temp.elem)
        temp = temp.next
    print()

# Driver code
arr = [5, 7, 6, 21, 3, 62, 213, 6, 121]
head = createList(arr)

print("Original Linked List:")
printLinkedList(head)
print("=====================")

new_head = removeAllOdd(head)

print("Linked List after removing odd elements:")
printLinkedList(new_head)
print("=====================")
```
