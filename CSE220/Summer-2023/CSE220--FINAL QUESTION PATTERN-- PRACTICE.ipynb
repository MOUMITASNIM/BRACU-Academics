Suppose you're given the head of a tree. Count the total number of the nodes that tree contains reccursively. (Following any other way will result in no marks) [10 Marks]

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def countNodes(root):
    if root is None:
        return 0
    return 1 + countNodes(root.left) + countNodes(root.right)


# Example tree:
#       1
#      / \
#     2   3
#    / \
#   4   5

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

total_nodes = countNodes(root)
print("Total nodes in the tree:", total_nodes)  # Output: 5
```

```python
#(b). To remove a node from a binary search tree using its predecessor (a node with the largest value in the left subtree)

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def findPredecessor(root):
    # Find the rightmost node in the left subtree
    current = root.left
    while current.right is not None:
        current = current.right
    return current

def deleteNode(root, key):
    if root is None:
        return root

    # Search for the node to be deleted
    if key < root.value:
        root.left = deleteNode(root.left, key)
    elif key > root.value:
        root.right = deleteNode(root.right, key)
    else:
        # Node with only one child or no child
        if root.left is None:
            return root.right
        elif root.right is None:
            return root.left

        # Node with two children, get the predecessor
        predecessor = findPredecessor(root)

        # Copy the predecessor's value to this node
        root.value = predecessor.value

        # Delete the predecessor
        root.left = deleteNode(root.left, predecessor.value)

    return root

# Example usage:
# Construct the tree
root = TreeNode(70)
root.left = TreeNode(30)
root.left.left = TreeNode(23)
root.left.right = TreeNode(45)
root.left.left.right = TreeNode(27)
root.right = TreeNode(85)
root.right.left = TreeNode(75)
root.right.right = TreeNode(90)
root.right.right.right = TreeNode(77)

# Delete a node using predecessor (e.g., delete node with value 70)
root = deleteNode(root, 70)
```

```python
#(c)To find the highest height (maximum depth) of a binary search tree, you can use the following Python program:

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def findMaxDepth(root):
    if root is None:
        return 0
    left_depth = findMaxDepth(root.left)
    right_depth = findMaxDepth(root.right)
    return max(left_depth, right_depth) + 1

# Example usage:
# Construct the tree (same as in part a)
root = TreeNode(70)
root.left = TreeNode(30)
root.left.left = TreeNode(23)
root.left.right = TreeNode(45)
root.left.left.right = TreeNode(27)
root.right = TreeNode(85)
root.right.left = TreeNode(75)
root.right.right = TreeNode(90)
root.right.right.right = TreeNode(77)

max_depth = findMaxDepth(root)
print("Highest height of the tree:", max_depth)
```

ANSWER--(a) :

a. To determine whether Tyrion's approach of creating a hash table of size 3 or Davos' approach of creating a hash table of size 5 is more optimized, you can consider the concept of load factor. The load factor is calculated as the ratio of the number of elements to the size of the hash table. A lower load factor generally results in fewer collisions and better performance.

Load Factor = (Number of Elements) / (Size of Hash Table)

Let's calculate the load factor for both approaches:

For Tyrion's approach (hash table of size 3):
Load Factor = (5 + 7 + 8 + 5 + 5) / 3 = 30 / 3 = 10

For Davos' approach (hash table of size 5):
Load Factor = (5 + 7 + 8 + 5 + 5) / 5 = 30 / 5 = 6

A lower load factor is generally better because it means there is more space available in the hash table to avoid collisions. Therefore, Davos' approach of creating a hash table of size 5 is more optimized in this case.

```python
#(b) answer:

#creating the hash table using Davos' hash function with linear probing to avoid collisions

class HashTable:
    def __init__(self, size):
        self.size = size
        self.keys = [None] * size
        self.values = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        while self.keys[index] is not None:
            # Linear probing to find the next available slot
            index = (index + 1) % self.size
        self.keys[index] = key
        self.values[index] = value

    def get(self, key):
        index = self.hash_function(key)
        while self.keys[index] is not None:
            if self.keys[index] == key:
                return self.values[index]
            # Linear probing to search for the key
            index = (index + 1) % self.size
        # Key not found
        return None

# Create a hash table of size 5
hash_table = HashTable(5)

# Insert data using Davos' hash function
data = [("Bangladesh", 5), ("India", 7), ("Pakistan", 8), ("Australia", 5), ("Sri Lanka", 5)]

for team, total_win in data:
    hash_table.insert(total_win, team)

# Example: Retrieve the team with 8 total wins
result = hash_table.get(8)
print("Team with 8 total wins:", result)
```
