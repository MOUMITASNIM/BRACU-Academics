```python
class Patient:
    def __init__(self, p_id, p_name, p_age, p_blood_group):
        self.p_id = p_id
        self.p_name = p_name
        self.p_age = p_age
        self.p_blood_group = p_blood_group
class WRM:
    def __init__(self):
        self.head = None

    def register_patient(self, p_id, p_name, p_age, p_blood_group):
        new_patient = Patient(p_id, p_name, p_age, p_blood_group)
        if self.head is None:
            self.head = new_patient
            new_patient.next = new_patient
            new_patient.prev = new_patient
        else:
            last_patient = self.head.prev
            last_patient.next = new_patient
            new_patient.prev = last_patient
            new_patient.next = self.head
            self.head.prev = new_patient

    def serve_patient(self):
        if self.head is None:
            print("No patients in the waiting room.")
            return

        serving_patient = self.head
        if self.head.next == self.head:  # Only 1 patient in the waiting room
            self.head = None
        else:
            self.head = self.head.next
            self.head.prev = serving_patient.prev
            serving_patient.prev.next = self.head
        print("Serving patient:", serving_patient.p_name)

    def cancel_all(self):
        if self.head is None:
            print("No patients in the waiting room.")
            return

        self.head = None
        print("All appointments are canceled. Doctor can go to lunch now.")

    def can_doctor_go_home(self):
        if self.head is None:
            return True
        return False

    def show_all_patients(self):
        if self.head is None:
            print("No patients in the waiting room.")
            return

        current_patient = self.head
        print("Waiting patients:")
        while True:
            print("ID:", current_patient.p_id, "Name:", current_patient.p_name)
            current_patient = current_patient.next
            if current_patient == self.head:
                break
def display_menu():
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Exit")

wrm = WRM()

while True:
    display_menu()
    choice = input("Enter your choice: ")

    if choice == "1":
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")
        wrm.register_patient(id, name, age, blood_group)
        print("Patient added successfully.")
    elif choice == "2":
        wrm.serve_patient()
    elif choice == "3":
        wrm.show_all_patients()
    elif choice == "4":
        if wrm.can_doctor_go_home():
            print("Yes, the doctor can go home.")
        else:
            print("No, the doctor cannot go home.")
    elif choice == "5":
        wrm.cancel_all()
    elif choice == "6":
        print("Closing the program.")
        break
    else:
        print("Invalid choice. Please try again.")
```
