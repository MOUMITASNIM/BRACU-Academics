```python
#TASK-1-(a)

def factl(n):
    if n == 0:
        return 1
    else:
        return n * factl(n-1)
print(factl(5))
```

```python
#TASK-1-(b)

def fibc(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibc(n-1) + fibc(n-2)
print(fibc(5))
```

```python
#TASK-1-(d)

def powerN(base, n):
    if n == 0:
        return 1
    else:
        return base * powerN(base, n-1)

print(powerN(5, 4))
```

```python
#TASK-2-(a)

def decimal_to_binary(n):
    if n == 0:
        return "0"
    elif n == 1:
        return "1"
    else:
        return decimal_to_binary(n // 2) + str(n % 2)

print(decimal_to_binary(5))
```

```python
#TASK-2-(b)

class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

def recurs_sum(node):
    if node == None:
        return 0
    return node.val + recurs_sum(node.next)

# Creating a linked list: [10,20,30,40]
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

total = recurs_sum(head)
print(total)
```

```python
#TASK-2-(c)

class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

def printReverse(node):
    if node is None:
        return
    printReverse(node.next)
    print(node.val)

# Creating linked list: 10 -> 20 -> 30 -> 40
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

print("Linked list's elements are shown in reversed order:")
printReverse(head)
```

```python
#TASK-3

def hocBuilder(height):
    if height == 0:
        return 0
    elif height == 1:
        return 8
    else:
        return hocBuilder(height - 1) + 5

print(hocBuilder(3))
```

```python
#TASK-5

import sys
sys.setrecursionlimit(10000)

class FinalQ:
    def print(self, array, idx):
        if idx < len(array):
            profit = self.calcProfit(array[idx])
            print(f"Investment: {array[idx]}; Profit: {profit}")
            self.print(array, idx + 1)

    def calcProfit(self, investment):
        if investment <= 100000:
            return self.profitHelper(investment - 25000, 4.5)
        else:
            return self.profitHelper(75000, 4.5) + self.profitHelper(investment - 100000, 8)

    def profitHelper(self, investment, percentage):
        if investment == 0:
            return 0
        else:
            return self.profitHelper(investment - 100, percentage) + (percentage / 100) * 100

# Tester
array = [25000, 100000, 250000, 350000]
f = FinalQ()
f.print(array, 0)
```
