```python
class Chitchat:
    def __init__(self, user_info: str, country: str = None):
        self.name, self.id, self.age, self.gender = user_info.split(',')
        self.country = country
        self.status = 'Online'
        self.friends = []
        self.friend_requests = []
        self.messages = []

    def AddFriend(self, friend, relation: str = None):
        if friend not in self.friends:
            friend.friend_requests.append((self, relation))
            return f'Friend request sent to {friend.name}'
        else:
            return f'{friend.name} is already your friend'

    def approveFriendReq(self):
        for friend, relation in self.friend_requests:
            self.friends.append((friend, relation))
            friend.friends.append((self, relation))
            print(f'{friend.name} and {self.name} are now friends')
        self.friend_requests = []

    def sendMessage(self, friend, message: str):
        if (friend, None) in self.friends:
            friend.messages.append((self, message))
            print(f'Message sent to {friend.name}')
        else:
            print(f'You are not friends with {friend.name}')

    def changeStatus(self):
        if self.status == 'Online':
            self.status = 'Offline'
        else:
            self.status = 'Online'
        print(f'{self.name} is now {self.status}')

    def showNetwork(self):
        print(f'{self.name}\'s Network:')
        for friend, relation in self.friends:
            if relation:
                print(f'{friend.name} ({relation})')
            else:
                print(friend.name)
print("Welcome to Chitchat! Create an account, have fun!")
Wahid = Chitchat("Wahid,00123,23,Male")
print("1======================================")
Maliha = Chitchat("Maliha,00124,25,Female", "Malaysia")
print("2======================================")
Sadik = Chitchat("Sadik,00125,26,Male", "Bangladesh")
print("3======================================")
print(Wahid.AddFriend(Sadik, "Brother"))
print("4======================================")
Wahid.sendMessage(Sadik, "Hi, bro. How are you?")
print("5======================================")
Sadik.approveFriendReq()
print("6======================================")
Wahid.sendMessage(Sadik, "Hi, bro. How are you?")
print("7======================================")
Wahid.changeStatus()
print("8======================================")
Sadik.sendMessage(Wahid, "Hello, I'm fine.")
print("9======================================")
Sadik.showNetwork()
print("10=====================================")
Lamia = Chitchat("Lamia,00126,26,Female")
print("11=====================================")
print(Sadik.AddFriend(Lamia))
print("12=====================================")
```
