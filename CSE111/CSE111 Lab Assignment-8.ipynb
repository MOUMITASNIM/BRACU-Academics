```python
#TASK-1
class RealNumber:
    def __init__(self, r=0):
        self.__realValue = r
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())

#Solution:
class ComplexNumber(RealNumber):
    def __init__(self, r=1,i=1):
        self.setRealValue(float(r))
        self._imaginaryValue=i
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())+"\n"+'ImaginaryPart: '+str(float(self._imaginaryValue))

# Execution commands
cn1 = ComplexNumber()
print(cn1)
print('---------') 
cn2 = ComplexNumber(5,7)
print(cn2)
```

```python
#TASK-2
class RealNumber:
    def __init__(self, number=0):
        self.number = number

    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number

    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number

    def __str__(self):
        return str(self.number)
#Solve:
class ComplexNumber(RealNumber):
    def __init__(self,number,imaginary_n):
        super().__init__(int(str(number)))
        self.imaginary_n=imaginary_n
    def __add__(self, anotherRealNumber):
        number=super().__add__(anotherRealNumber)
        imaginary_n = self.imaginary_n + anotherRealNumber.imaginary_n
        return ComplexNumber(number,imaginary_n)
    def __sub__(self, anotherRealNumber):
        number=super().__sub__(anotherRealNumber)
        imaginary_n = self.imaginary_n-anotherRealNumber.imaginary_n
        return ComplexNumber(number,imaginary_n)
    def __str__(self):
        if self.imaginary_n>0:
            return str(self.number)+" + "+str(self.imaginary_n)+"i"
        else:
             return str(self.number)+" - "+str(abs(self.imaginary_n))+"i"

#execution Commands:
r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1 + r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
#TASK-5
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

#Solution:
class  ScienceExam(Exam):
    def __init__(self,marks,time,*subjects):
        super().__init__(marks)
        self.time=time
        self.subjects=list(subjects)
        self.parts=2+len(subjects)
    def __str__(self):
        return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: {self.parts}"
    def examSyllabus(self):
        subj="Maths , English"
        for e in self.subjects:
                subj+=" , "+e
        return subj
    def examParts(self):
        subj="Part 1 - Maths\nPart 2 - English\n"
        count_s=3
        for element in self.subjects:
            if element!=self.subjects[-1]:
                subj+="Part "+str( count_s)+" - "+element+"\n"
            else:
                 subj+="Part "+str( count_s)+" - "+element
            count_s+=1
        return subj
# Execution commands  
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture =ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#TASK-6

class Shape3D:
    pi = 3.14159
    def __init__(self, name = 'Default', radius = 0):
        self._area = 0
        self._name = name
        self._height = 'No need'
        self._radius = radius
    def calc_surface_area(self):
        return 2 * Shape3D.pi * self._radius
    def __str__(self):
        return "Radius: "+str(self._radius)

#Solution:
class Sphere(Shape3D):
    def __init__(self, name, radius):
        print("Shape name: Sphere, Area Formula: 4 * pi * r* r")
        super().__init__(name,radius)
        self._area=0.0
    def calc_surface_area(self):
        self._area= 4*Shape3D.pi*self._radius**2
    def __str__(self):
        return "Area: "+str(self._area)
class  Cylinder(Shape3D):
    def __init__(self, name, radius,height):
        print("Shape name: Cylinder, Area Formula: 2 * pi *r * (r + h)")
        super().__init__(name,radius)
        self._height=height
        self._area=0.0
    def calc_surface_area(self):
        self._area=2*Shape3D.pi*self._radius*(self._radius+self._height)
    def __str__(self):
        return "Area: "+str(self._area)
    
# Execution commands    
sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
#TASK-7
class PokemonBasic:
    def __init__(self, name = 'Default', hp = 0,weakness = 'None', type = 'Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type
    def get_type(self):
        return 'Main type: ' + self.type
    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'
    def __str__(self):
        return "Name: " + self.name + ", HP: " +str(self.hit_point) + ", Weakness: " + self.weakness
        print('\n------------Basic Info:--------------')

#Solution:
class PokemonExtra(PokemonBasic):
    def __init__(self,name = 'Default', hp = 0,weakness = 'None', type = 'Unknown',secondary_typ=None,other_move=None):
        super().__init__(name , hp,weakness, type )
        self.secondary_typ=secondary_typ
        self.other_move=other_move
        if secondary_typ!=None:
             super().get_type()
        if other_move!=None:
            super().get_move
    def get_type(self):
        if self.secondary_typ!=None:
            return f'Main type:{self.type},Secondary type:{self.secondary_typ}'
        else:
            return 'Main type: ' + self.type
    def get_move(self):
        if self.other_move!=None:
            m,n=self.other_move 
            return f'Basic move:Quick Attack \nOther move:{m},{n}'
        else:
            return 'Basic move: ' + 'Quick Attack'
               
# Execution commands 
print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water','Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water','Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#TASK-9
class Pokemon:
    def __init__(self, p):
        self.pokemon = p
        self.pokemon_type = "Needs to be set"
        self.pokemon_weakness = "Needs to be set"

    def kind(self):
        return self.pokemon_type

    def weakness(self):
        return self.pokemon_weakness

    def what_am_i(self):
        print("I am a Pokemon.")

#solution:
class Pikachu(Pokemon):
    def __init__(self, name="Pikachu", type="Electric", weakness="Ground"):
        super().__init__(name)
        self.pokemon_type = type
        self.pokemon_weakness = weakness

    def what_am_i(self):
        super().what_am_i()
        print(f"I am {self.pokemon}")


class Charmander(Pokemon):
    def __init__(self, name="Charmander", type="Fire", weakness="Water, Ground and Rock"):
        super().__init__(name)
        self.pokemon_type = type
        self.pokemon_weakness = weakness

    def what_am_i(self):
        super().what_am_i()
        print(f"I am {self.pokemon}")

#execution_commands
pk1 = Pikachu()
print("Pokemon:", pk1.pokemon)
print("Type:", pk1.kind())
print("Weakness:", pk1.weakness())
pk1.what_am_i()
print("========================")
c1 = Charmander()
print("Pokemon:", c1.pokemon)
print("Type:", c1.kind())
print("Weakness:", c1.weakness())
c1.what_am_i()
```

```python
#TASK-10
class Department:
    def __init__(self, s):
        self.semester = s
        self.name = "Default"
        self.id = -1

    def student_info(self):
        print("Name:", self.name)
        print("ID:", self.id)

    def courses(self, c1, c2, c3):
        print("No courses Approved yet!")

#solution:
class CSE(Department):
    def __init__(self, name, ID, semester):
        super().__init__(semester)
        self.name = name
        self.id = ID

    def student_info(self):
        super().student_info()
        print(f"Courses Approved to this CSE student in {self.semester} semester: ")

    def courses(self, c1, c2, c3):
        print(f"{c1}\n{c2}\n{c3}")


class EEE(Department):
    def __init__(self, name, ID, semester):
        super().__init__(semester)
        self.name = name
        self.id = ID

    def student_info(self):
        super().student_info()
        print(f"Courses Approved to this EEE student in {self.semester} semester: ")

    def courses(self, c1, c2, c3):
        print(f"{c1}\n{c2}\n{c3}")

#execution_commands
s1 = CSE("Rahim", 16101328,"Spring2016")
s1.student_info()
s1.courses("CSE110", "MAT110", "ENG101")
print("==================")
s2 = EEE("Tanzim", 18101326, "Spring2018")
s2.student_info()
s2.courses("Mat110", "PHY111", "ENG101")
print("==================")
s3 = CSE("Rudana", 18101326, "Fall2017")
s3.student_info()
s3.courses("CSE111", "PHY101", "MAT120")
print("==================")
s4 = EEE("Zainab", 19201623, "Summer2019")
s4.student_info()
s4.courses("EEE201", "PHY112", "MAT120")
```
