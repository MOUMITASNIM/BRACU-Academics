```python
#task_4

class Color:
  def __init__(self,clr):
    self.clr=clr
  def __add__(self,other):
    if self.clr=="red" and other.clr=="yellow" or self.clr=="yellow" and other.clr=="red":
      return Color("Orange")
    elif self.clr=="red" and other.clr=="blue" or self.clr=="blue" and other.clr=="red":
       return Color("Violet")
    elif self.clr=="yellow" and other.clr=="blue" or self.clr=="blue" and other.clr=="yellow":
       return Color("Green")
       


#Driver_code    
    
C1 = Color(input("First Color: ").lower())
C2 = Color(input("Second Color: ").lower())
C3 = C1 + C2
print("Color formed:", C3.clr)
```

```python
#task_5
import math
class Circle:
  def __init__(self,rad):
    self.__rad=rad
  def setRadius(self):
     self.__rad=rad
  def getRadius(self):
    return self.__rad
  def area(self):
    area=math.pi*self.__rad**2
    return area
  def __add__(self,other):
    self.__rad+=other.__rad
    return Circle(self.__rad)

    
#driver_code
c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" ,c1.area())
 
c2 = Circle(5)
print("Second circle radius:" ,c2.getRadius())
print("Second circle area:" ,c2.area())
 
c3 = c1 + c2 
print("Third circle radius:" ,c3.getRadius())
print("Third circle area:" ,c3.area())
```

```python
#task_6
class Triangle:
    def __init__(self, t_base=None, t_height=None):
        self.__t_base = t_base
        self.__t_height = t_height

    def getBase(self):
        return self.__t_base

    def setBase(self, t_base):
        self.__t_base = t_base
        return self.__t_base

    def getHeight(self):
        return self.__t_height

    def setHeight(self, t_height):
        self.__t_height = t_height
        return self.__t_height

    def area(self):
        self.area = 0.5 * self.__t_base * self.__t_height
        return self.area

    def __sub__(self, other):
        self.new_base = self.__t_base - other.__t_base
        self.new_height = self.__t_height - other.__t_height
        return Triangle(self.new_base, self.new_height)

#Driver_code    
t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
 
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
 
t3 = t1 - t2 
print("Third Triangle Base:" , t3.getBase())
print("Third Triangle Height:" , t3.getHeight())
print("Third Triangle area:" ,t3.area())
```

```python
#task_8
class Coordinates:
    def __init__(self,m,n ):
        self.m = m
        self.n =n

    def detail(self):
        return (self.m, self.n)

    def __sub__(self, other):
        a = self.m - other.m
        b = self.n - other.n
        return Coordinates(a, b)

    def __mul__(self, other):
        c = self.m * other.m
        d = self.n * other.n
        return Coordinates(c, d)

    def __eq__(self, other):
        if self.m == other.m:
            return "The calculated coordinates are the same."
        else:
            return "The calculated coordinates are NOT the same."

#Driver code 

p1 = Coordinates(int(input()),int(input()))
p2 = Coordinates(int(input()),int(input()))

p4 = p1 - p2 
print(p4.detail())

p5 = p1 * p2
print(p5.detail())

point_check = (p4 == p5)
print(point_check)
```
