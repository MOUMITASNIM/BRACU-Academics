```python
#TASK-2-HOMEWORK

class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

#Solution:
class  ScienceExam(Exam):
    def __init__(self,marks,time,*subjects):
        super().__init__(marks)
        self.time=time
        self.subjects = subjects[:]
        self.parts=len(subjects)+2
    def __str__(self):
        return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: {self.parts}"
    def examSyllabus(self):
        subj="Maths , English"
        for e in self.subjects:
                subj+=" , "+e
        return subj
    def examParts(self):
        subj="Part 1 - Maths\nPart 2 - English\n"
        count_s=3
        for element in self.subjects:
            if element!=self.subjects[-1]:
                subj+="Part "+str( count_s)+" - "+element+"\n"
            else:
                 subj+="Part "+str( count_s)+" - "+element
            count_s+=1
        return subj
# Execution commands
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture =ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#TASK-3-HOMEWROK

class PokemonBasic:
    def __init__(self, name = 'Default', hp = 0,weakness = 'None', type = 'Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type
    def get_type(self):
        return 'Main type: ' + self.type
    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'
    def __str__(self):
        return "Name: " + self.name + ", HP: " +str(self.hit_point) + ", Weakness: " + self.weakness
        print('\n------------Basic Info:--------------')

#Solution:
class PokemonExtra(PokemonBasic):
    def __init__(self,name = 'Default', hp = 0,weakness = 'None', type = 'Unknown',secondary_typ=None,other_move=None):
        super().__init__(name , hp,weakness, type )
        self.secondary_typ=secondary_typ
        self.other_move=other_move
        if secondary_typ!=None:
             super().get_type()
        if other_move!=None:
            super().get_move
    def get_type(self):
        if self.secondary_typ!=None:
            return f'Main type:{self.type},Secondary type:{self.secondary_typ}'
        else:
            return 'Main type: ' + self.type
    def get_move(self):
        if self.other_move!=None:
            m,n=self.other_move
            return f'Basic move:Quick Attack \nOther move:{m},{n}'
        else:
            return 'Basic move: ' + 'Quick Attack'

# Execution commands
print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water','Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water','Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#TASK-4-CLASSWORK

class Cakes:
    order_list = {}
    feedbacks = {}
    def __init__(self,flavor,weight,sweetness="Moderate",cream_type="Whipped Cream"):
        self.flavor = flavor
        self.weight = weight
        self.sweetness = sweetness
        self.cream_type = cream_type
        self.price = weight*1.2
        if f"{flavor} Cake {weight}gm" not in Cakes.order_list:
            Cakes.order_list[f"{flavor} Cake {weight}gm"] = 1
        else:
            Cakes.order_list[f"{flavor} Cake {weight}gm"] += 1

    def add_customization(self, sweetness=None, cream_type=None):
       if sweetness is not None:
        self.sweetness = sweetness
       if cream_type is not None:
        self.cream_type = cream_type


    def cake_details(self):
        print(f"Flavor: {self.flavor} Cake, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweetness} sugar, Cream Type: {self.cream_type}")
        print(f"Price: {self.price} Taka")

    @classmethod
    def give_feedbacks(cls,cake,feedback):
        if cake not in cls.feedbacks:
            cls.feedbacks[cake] = [feedback]
        else:
            cls.feedbacks[cake].append(feedback)
        print("Thanks for your valuable feedback!")

    @classmethod
    def show_feedbacks(cls):
        print(cls.feedbacks)

class Cheese_Cakes(Cakes):
    def __init__(self,flavor,weight,cake_type="baked",sweetness="Moderate",cream_type="Cream Cheese"):
        super().__init__(flavor,weight,sweetness,cream_type)
        self.cake_type = cake_type
        self.price = weight*1.5
        if f"{flavor} Cheese Cake {weight}gm" not in Cakes.order_list:
            Cakes.order_list[f"{flavor} Cheese Cake {weight}gm"] = 1
        else:
            Cakes.order_list[f"{flavor} Cheese Cake {weight}gm"] += 1

    def add_customization(self,sweetness=None,cream_type=None):
        print("Sorry! No customization available for cheese cakes.")

    def cake_details(self):
        print(f"Flavor: {self.flavor} Cheese Cake, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweetness} sugar, Cream Type: {self.cream_type}")
        print(f"Cake Type:{self.cake_type}, Price: {self.price} Taka")

    @classmethod
    def give_feedbacks(cls,cake,feedback):
        super().give_feedbacks(cake,feedback)
        print("You will get 10% discount for your next purchase!")

#driver_code
order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake","Very Delicious")
Cakes.give_feedbacks("Chocolate Cake","Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake","Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake","Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()
```
