```python
#QUESTION_3


class Sphere:
    def __init__(self, sphr_n, rad=1, color="White",pi= 3.1416):
        self.sphr_n = sphr_n
        self.rad = rad
        self.color = color
        self.s_vol = (4/3) * 3.1416 * (self.rad ** 3)

    def printDetails(self):
        print(f'Sphere ID:{self.sphr_n}')
        print(f'Color:{self.color} ')
        print(f'Volume: {self.s_vol}')


    def merge_sphere(self,*s_spheres):
      print("Spheres are being merged")

      for i in s_spheres:
         self.s_vol += i.s_vol

         if self.color == i.color:
            continue

         else:
          self.color="Mixed Color"



#driver_code
sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```
