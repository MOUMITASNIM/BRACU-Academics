```python
class Movie:
    def __init__(self, name, genre, duration):
        self.name = name
        self.genre = genre
        self.duration = duration

    def movieInfo(self):
        return f"Name: {self.name}, Genre: {self.genre}, Duration: {self.duration} minutes"

    @classmethod
    def createMovie_fromString(cls, movie_string):
        name, genre, duration = movie_string.split('-')
        return cls(name, genre, int(duration))


class StarCinema:
    all_branch_info = {}

    def __init__(self, branch_name):
        self.branch_name = branch_name
        self.movie_list = []
        StarCinema.all_branch_info[branch_name] = self.movie_list

    def addMovies(self, *movie_objects):
        for movie in movie_objects:
            if movie not in self.movie_list:
                self.movie_list.append(movie)

    def removeMovie(self, movie_object):
        if movie_object in self.movie_list:
            self.movie_list.remove(movie_object)


    @classmethod
    def check(cls, movie_name):
       found = False
       for branch_name in cls.all_branch_info:
        movie_list = cls.all_branch_info[branch_name]
        for movie in movie_list:
            if movie.name == movie_name:
                print(f"Movie '{movie_name}' is streaming at {branch_name} branch.")
                print(movie.movieInfo())
                found = True
        if not found:
           print(f"Movie '{movie_name}' is not streaming in any branch.")

    @classmethod
    def showAllBranchInfo(cls):
      for branch_name in cls.all_branch_info:
        movie_list = cls.all_branch_info[branch_name]
        print(f"Branch Name: {branch_name}")
        mov_num = 1  # Manual index counter
        for movie in movie_list:
            print(f"Movie No: {mov_num}")
            print(movie.movieInfo())
            print('-' * 30)
            mov_num += 1  # Increment the counter



#driver code
# Creating Movie objects from string data
movies = [
    Movie.createMovie_fromString('Amar Ache Jol-Drama-150'),
    Movie.createMovie_fromString('Bumblebee-Adventure-168'),
    Movie.createMovie_fromString('Feluda-Detective-149'),
    Movie.createMovie_fromString('Pother Pachali-Drama-90')
]

# Creating StarCinema branch objects
gulshan_branch = StarCinema('Gulshan')
banani_branch = StarCinema('Banani')

# Adding movies to branches
gulshan_branch.addMovies(movies[0], movies[1])
banani_branch.addMovies(movies[0], movies[2], movies[3])

# Showing initial branch information
print("Previous Branch Information:")
StarCinema.showAllBranchInfo()

# Checking which branches a movie is streaming in
StarCinema.check('Amar Ache Jol')
StarCinema.check('Feluda')
StarCinema.check('Bumblebee')
StarCinema.check('Pother Pachali')

# Removing a movie from a branch
gulshan_branch.removeMovie(movies[0])

# Showing updated branch information
print("\nNewly Updated Branch Info:")
StarCinema.showAllBranchInfo()
```
