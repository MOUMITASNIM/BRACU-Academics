```python
# Design the Monster class with necessary properties so that the given output is produced.
# [Hint:
# 1. Let's say there are 2 monsters: A and B.
# -A can attack B if A is alive.
# -If A is alive, it can only attack B if B is alive.
# -If A's power is greater than that of B, only then A can defeat B;
# Otherwise, A will be defeated by B.
# - Once a monster gets defeated, it is considered to be dead.
# 2. The variable "monsterCount" keeps track of the number of monsters
# alive. So, update it accordingly.]
class Monster:
  monsterCount=0
  def __init__(self,n,p,a=True):
    self.name=n
    self.pow=p
    self.al=a
    Monster.monsterCount+=1
  def get_details(self):
    return (f"Name:{self.name} \nPower:{self.pow} \nAlive:{self.al}")  
  def attack(self,*m):
    if len(m)==0:
      print("No monsters to attack")
    else:
      for i in m:
        if self.al==True and i.al==True:
         if self.pow>i.pow:
          print(f"Attack successful. {self.name} defeated {i.name}")
          i.al=False
          Monster.monsterCount-=1
         else:
          print(f"Attack unsuccessful. {self.name} was defeated by {i.name}")
          self.al=False 
          Monster.monsterCount-=1 
        elif self.al==True and i.al==False:
          print(f"Cannot attack {i.name}. It's not alive")
        elif self.al==False:
          print(f"{self.name} is not alive to attack.")  


monster1 = Monster('Godzilla', 40)
monster2 = Monster('Hydra', 30)
monster3 = Monster('KingKong', 50)
print(f"Number of monsters alive:{Monster.monsterCount}")
print('1--------------------------------')
print(monster1.get_details())
print('2--------------------------------')
print(monster2.get_details())
print('3--------------------------------')
print(monster3.get_details())
print('4--------------------------------')
monster1.attack()
print('5--------------------------------')
monster1.attack(monster2)
print('6--------------------------------')
monster1.attack(monster2, monster3)
print('7--------------------------------')
print(f"Number of monsters alive:{Monster.monsterCount}")
print('8--------------------------------')
print(monster2.get_details())
print('9--------------------------------')
monster2.attack(monster1)

# Number of monsters alive:3
# 1--------------------------------
# Name:Godzilla
# Power:40
# Alive:True
# 2--------------------------------
# Name:Hydra
# Power:30
# Alive:True
# 3--------------------------------
# Name:KingKong
# Power:50
# Alive:True
# 4--------------------------------
# No monsters to attack
# 5--------------------------------
# Attack successful.Godzilla defeated
# Hydra.
# 6--------------------------------
# Cannot attack Hydra. It's not alive.
# Attack unsuccessful. Godzilla was
# defeated by KingKong.
# 7--------------------------------
# Number of monsters alive:1
# 8--------------------------------
# Name:Hydra
# Power:30
# Alive:False
# 9--------------------------------
# Hydra is not alive to attack.
```
