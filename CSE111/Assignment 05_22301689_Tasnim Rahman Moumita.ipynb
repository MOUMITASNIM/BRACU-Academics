```python
class Book:
    def __init__(self, title, author, genre):
        self.__title = title
        self.__author = author
        self.__genre = genre
        self.__available = True
        self.__borrower = None

    def set_title(self, title):
        self.__title = title

    def get_title(self):
        return self.__title

    def set_author(self, author):
        self.__author = author

    def get_author(self):
        return self.__author

    def set_genre(self, genre):
        self.__genre = genre

    def get_genre(self):
        return self.__genre

    def set_availability(self, available):
        self.__available = available

    def get_availability(self):
        return self.__available

    def set_borrower(self, borrower):
        self.__borrower = borrower

    def get_borrower(self):
        return self.__borrower

    def display_info(self):
        print(f"Title: {self.title}\nAuthor: {self.author}\nGenre: {self.genre}\nAvailable: {self.available}\nBorrowed by: {self.borrower}\n-----")


class LibraryMember:
    def __init__(self, member_id, name):
        self.__member_id = member_id
        self.__name = name
        self.__borrowed_books = []

    def set_member_id(self, member_id):
        self.__member_id = member_id

    def get_member_id(self):
        return self.__member_id

    def set_name(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

    def borrow_book(self, book):
        if book.get_availability():
            book.set_availability(False)
            book.set_borrower(self)
            self.__borrowed_books.append(book)
            print(f" '{book.get_title()}' borrowed by: {self.get_name()}.")
        else:
            print(f"The book '{book.get_title()}'not available")

    def return_book(self, book):
        if book in self.__borrowed_books:
            book.set_availability(True)
            book.set_borrower(None)
            self.__borrowed_books.remove(book)
            print(f"'{book.get_title()}' returned by: {self.get_name()}.")
        else:
            print(f"{self.get_name()} did not borrow the book '{book.get_title()}'.")

    def display_borrowed_books(self):
        if len(self.__borrowed_books) == 0:
            print(f"{self.get_name()} did not borrow.")
            return
        print(f"{self.get_name()} has borrowed the following books:")
        for book in self.__borrowed_books:
            print(f"- {book.get_title()}")


class Library:
    def __init__(self):
        self.books_available = []
        self.library_members = []

    def add_book(self, book):
      if type(book).__name__ == 'Book':
          self.books_available.append(book)
          print(f"The book '{book.get_title()}' has been added to the library collection.")
      else:
          print("Invalid input.")

    def add_library_member(self, member):
      if type(member).__name__ == 'LibraryMember':
          self.library_members.append(member)
          print(f"{member.get_name()} has been added as a library member.")
      else:
          print("Invalid input.")

    def display_book_list(self):
      if len(self.books_available) == 0:
          print("No book available.")
          return
      print("There are the following books available:")
      for book in self.books_available:
          print(f"- {book.get_title()}")

    def display_library_members(self):
      if len(self.library_members) == 0:
          print("There are no members.")
          return
      print("There are the following members:")
      for member in self.library_members:
          print(f"- {member.get_name()}")
      print(f"- {member.get_name()}")




# Create a Library object
my_library = Library()
print("-----------------------------------------------")

# Create some Book objects
print("----------------------------------------------")
book1 = Book("Deyal", "Humayun Ahmed", "Historical")
book2 = Book("Kalbela", "Samaresh Majumder", "Political")
book3 = Book("Tukunjil", "Muhammad Jafar Iqbal", "Science Fiction")

# Add books to the library
print("---------------------------------------------")
my_library.add_book(book1)
my_library.add_book(book2)
my_library.add_book(book3)

# Display the list of books in the library
print("--------------------------------------------")
my_library.display_book_list()

# Create some LibraryMember objects
print("--------------------------------------------")
member1 = LibraryMember("Member-01", "TASNIM")
member2 = LibraryMember("Member-02", "SAZZAD")


# Add members to the library
print("--------------------------------------------")
my_library.add_library_member(member1)
my_library.add_library_member(member2)

# Display the list of library members
print("-------------------------------------------")
my_library.display_library_members()

# Borrow some books
print("-------------------------------------------")
member1.borrow_book(book1)
member1.borrow_book(book2)
member2.borrow_book(book3)

# Display borrowed books
print("-------------------------------------------")
member1.display_borrowed_books()
member2.display_borrowed_books()

# Return some books
print("-------------------------------------------")
member1.return_book(book1)
member2.return_book(book3)

# Display borrowed books again
print("-------------------------------------------")
member1.display_borrowed_books()
member2.display_borrowed_books()
```
