```python
#task_2
class Shape:
  def __init__(self,a,b,c):
   self.a=a
   self.b=b
   self.c=c
   self.triangle=1/2* self.b*self.c
   self.square= self.b**2
   self.rectangle=self.b*self.c
   self.rhombus=1/2* self.b*self.c
  def area(self):
    if (self.a=="Triangle"):
     print(f"Area:{ self.triangle}")
    elif (self.a=="Square"):
      print(f"Area:{self.square}")
    elif (self.a=="Rectangle"):
      print(f"Area:{ self.rectangle}")
    elif (self.a=="Rhombus"):
      print(f"Area:{ self.rhombus}")
    else:
      print("Area: Shape unknown")

#driver_code
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#task_3
class Calculator:
  def __init__(self):
    print("Calculator is ready!")
    self.p=None
    self.q=None
    self.op="+"
    self.val=None

  def calculate(self,p,q,op):
    self.p=p
    self.q=q
    self.op=op

    if(op=="+"):
      val=p+q
      p=val
    elif(op=="-"):
      val=p-q
      p=val
    elif(op=="*"):
      val=p*q
      p=val
    elif(op=="/"):
      val=p/q
      p=val
    self.val=val
    return val
  def showCalculation(self):
   print(self.p,self.op,self.q,"=",val)

#driver code
c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#task_4
class Programmer:
    def __init__(self, name, language, experience):
        self.name = name
        self.language = language
        self.experience = experience
        self.flag = True

    def addExp(self, additional_exp):
        self.flag = False
        print(f"Updating experience of {self.name}")
        self.experience += additional_exp

    def printDetails(self):
        if self.flag:
            print("Horray! A new programmer is born")
        print(f"Name: {self.name}")
        print(f"Language: {self.language}")
        print(f"Experience: {self.experience} years")

#driver code
p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print("--------------------------")
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print("--------------------------")
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
#task_6
class Customer:
    def __init__(self,c_name):
        self.c_name = c_name

    def greet(self, c_name=None):
        if c_name != None:
            print(f"Hello {c_name}!")
        else:
            print("Hello!")

    def purchase(self, *item):
        print(f"{self.c_name}, you purchased {len(item)} items(s):")
        for i in item:
            print(i)

#driver code
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#task_8
class Student:
    def __init__(self, name, ID, dept="CSE"):
        self.name = name
        self.ID = ID
        self.dept = dept

    def dailyEffort(self, hours):
        self.hours = hours

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.ID}")
        print(f"Department: {self.dept}")
        print(f"Daily Effort: {self.hours} hour(s)")

        if self.hours <= 2:
            print("Suggestion: Should give more effort!")
        elif self.hours <= 4:
            print("Suggestion: Keep up the good work!")
        else:
            print("Suggestion: Execellent! Now motivate others.")

#driver code
harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#task_9
class Batsman:
    def __init__(self, *args):
        if len(args)==2:
            runs,balls=args
        else:
            player,runs,balls=args

        if len(args)!=2:
            self.name=player
        else:
            self.name="New Batsman"

        self.runs=runs
        self.balls=balls

    def setName(self, name):
        self.name=name

    def battingStrikeRate(self):
        return (self.runs)/(self.balls)*100

    def printCareerStatistics(self):
        print(f"Name: {self.name}")
        print(f"Runs Scored: {self.runs}, Balls Faced: {self.balls}")

#driver code
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#task_10
class Author:
    def __init__(self, name=None, *books):
        self.name=name
        if name==None:
            self.name="Default"

        booklist=""
        if len(books)>0:
            for i in books:
                booklist+=i+"\n"
            self.books=booklist

    def changeName(self, new_name):
        self.name=new_name

    def addBooks(self, *books):
        booklist= ""
        for i in books:
            booklist+=i + "\n"

        self.books=booklist

    def printDetails(self):
        print(f"Author Name: {self.name}")
        print("------------")
        print(f"List of Books:")
        print(self.books)

#driver code
auth1 = Author("Humayun Ahmed")
auth1.addBooks("Deyal", "Megher Opor Bari")
auth1.printDetails()
print("===================")
auth2 = Author()
print(auth2.name)
auth2.changeName("Mario Puzo")
auth2.addBooks("The Godfather", "Omerta", "The Sicilian")
print("===================")
auth2.printDetails()
print("===================")
auth3 = Author("Paolo Coelho", "The Alchemist", "The Fifth Mountain")
auth3.printDetails()
```

```python
#task_12
class Account:
    def __init__(self, name=None, balance=None):
        self.name=name
        self.balance=balance

        if (self.name==None):
            self.name="Default Account"

        if (self.balance==None):
            self.balance=0.0

    def withdraw(self, amount):
        var=self.balance - amount
        if (var<=3070.0):
            print(
                "Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum."
            )
        else:
            print(f"Withdraw successful! New balance is: {var}")

    def details(self):
        return f"{self.name}\n{self.balance:.1f}"

#driver_code
a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
#task_13
class StudentDatabase:
  def __init__(self,name,st):
    self.name=name
    self.st=st
    self.grades={}
  def calculateGPA(self,S_List,session):
    total_grade=0
    subjects=list()
    for subject in S_List:
      s_Name=subject[:subject.find(':')]
      s_Grade=float(subject[subject.find(':')+1:])
      total_grade+= s_Grade
      subjects.append(s_Name)

      GPA=round(total_grade/len(S_List),2)
      self.grades[session]={tuple(subjects):GPA}
  def printDetails(self):
    print(f"Name: {self.name}")
    print(f"ID: {self.st}")

    for session in self.grades:
      print(f"Courses taken in {session}:")

      for item in self.grades[session].items():
        for subject in item[0]:
          print(subject)
        print(f"GPA:{item[1]}")

#driver code

# Do not change the following lines of code.

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
