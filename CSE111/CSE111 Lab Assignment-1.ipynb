```python
#task_4
def replace_domain(email,new_domain,old_domain='kaaj.com'):

    prev_email = email.split('@')
    if prev_email[1] == old_domain:
        result = prev_email[0] + '@' + new_domain
        print ("Changed:",result)
    else:
        print ("Unchanged:",email)
    

replace_domain ('alice@kaaj.com','sheba.xyz','kaaj.com')
replace_domain ('bob@sheba.xyz','sheba.xyz')
```

```python
#task_5
def check_palindrome(word):
  new_word = ''
  old_word = ''
  
  for i in range(len(word)):
    if word[i] == ' ':
      continue
    else:
      old_word += word[i]

  for i in range(-1,-(len(old_word)+1),-1):
    new_word += old_word[i]

  if (old_word == new_word):
    print("Palindrome")
  else:
    print("Not a palindrome")

pali_input = input()
check_palindrome(pali_input)
```

```python
#task_6
def upper(sentence):
  symbol= '.?!'
  new_sen = ''
  space_n = 0
  punc_n = 0

  for i in range(len(sentence)):
    if (i == 0):
      new_sen += sentence[i].upper()
      continue
    else:
      if (sentence[i] in symbol):
        punc_n +=1
        new_sen+= sentence[i]
      elif (sentence[i] == " "):
        space_n +=1
        if (space_n > 1):
          space_n = 0
        new_sen+= sentence[i]
      
      elif (punc_n == 1 and space_n == 1):
        new_sen+= sentence[i].upper()
        punc_n = 0
        space_n = 0
      elif (space_n == 1 and sentence[i] == "i"):
          if (sentence[i+1] == ' '):
            new_sen+= sentence[i].upper()
          else:
            new_sen+= sentence[i]
          space_n = 0
          punc_n = 0
      else:
        new_sen+= sentence[i]
        space_n = 0
        punc_n = 0

  return new_sen

us_input = input()
print(upper(us_input))
```
