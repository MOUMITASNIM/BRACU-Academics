Given integral, $\int_{0}^{2}\sqrt{4-x^2}dx$




(a)Using SymPy, calculate  I  upto  5  decimal  points .


```python
#a)Solution:

import sympy as sp

x = sp.symbols('x')
I_sy = sp.integrate(sp.sqrt(4 - x**2), (x, 0, 2))                                 #computes the integral of the function
print("The integral has been computed using SymPy: I =",round(I_sy.evalf(), 5))   #Rounds the output upto 5 decimal points and
                                                                                  #evaluates the symbolic integral to a numerical value.
```

(b)Using trapezoidal method and varying step size h, calculate I accurate upto 5
decimal points. You can use NumPy but cannot use SymPy or SciPy.

```python
#(b)Solution:

import numpy as np

def f(x):
    return np.sqrt(4 - x**2)

def trapezoid_method(p, q, n):
    h = (q - p) / n
    x = np.linspace(p, q, n+1)               #Generating an array of equally spaced points between p and q using linspace().
    y = f(x)
    area = h * (y[0] + y[-1]) / 2
    area += h * np.sum(y[1:-1])
    return area

I_tr = trapezoid_method(0, 2, 9899)
print("The integral has been computed using Trapezoidal method: I =", round(I_tr, 5))   #Rounds the output upto 5 decimal points.
```

(c)Let’s calculate I using Monte Carlo. Let N be the total number of random
points inside by rectangle containing the desired curve. I(N) is the calculated
area under the curve using Monte Carlo.
Using Matplotlib, plot I(N) vs N, where N = 100, 200, 300, ...., 2000.
Use the seed = last two digit of ID × MAT 120 Section No.
(i.e., if you are in section 5, with ID 17105412, then seed = 12 × 5 = 60.)

```python
#(c)Solution:

from numpy import *
from matplotlib.pyplot import *
from numpy.random import *

def f(x):
    return np.sqrt(4 - x**2)


def monte_carlo_method(N):

    x = 2 *rand(N)
    y = 2 *rand(N)
    count = 0

    for i in range(len(y)):
      if y[i] < f(x[i]):
        count += 1

    area = 2 * 2              #Calculating the area of the rectangle (here, it’s a fixed rectangle with dimensions 2x2)
    return area * count / N   #Returning the estimated integral using the Monte Carlo method.

seed = 89 * 17                # My ID:22301689, My MAT120 section: 17

N = []
c = 100
while c <= 2000:
    N.append(c)
    c += 100

I = [monte_carlo_method(N) for N in N]

plot(N, I)
xlabel('Number of random points (N)')
ylabel('Monte Carlo Integral estimate (I)')
title('Monte Carlo Integration')
grid()
show()
```
